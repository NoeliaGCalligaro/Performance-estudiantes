```python
import pandas as pd
import numpy as np
import os
import zipfile

# Reextraer y cargar los datos



file_path = 'data/student_habits_performance.csv'
df = pd.read_csv(file_path)



# Copia para transformaci√≥n
df_etl = df.copy()

# Codificaci√≥n binaria
binary_map = {
    'Yes': 1,
    'No': 0
}
df_etl['part_time_job'] = df_etl['part_time_job'].map(binary_map)
df_etl['extracurricular_participation'] = df_etl['extracurricular_participation'].map(binary_map)

# Codificaci√≥n ordinal
diet_map = {'Poor': 0, 'Fair': 1, 'Good': 2}
edu_map = {'High School': 0, 'Bachelor': 1, 'Master': 2, 'PhD': 3}
internet_map = {'Poor': 0, 'Average': 1, 'Good': 2}
df_etl['diet_quality'] = df_etl['diet_quality'].map(diet_map)
df_etl['parental_education_level'] = df_etl['parental_education_level'].map(edu_map)
df_etl['internet_quality'] = df_etl['internet_quality'].map(internet_map)

# Codificaci√≥n nominal (One-Hot) para g√©nero
df_etl = pd.get_dummies(df_etl, columns=['gender'], prefix='gender')

# Creaci√≥n de variables nuevas
df_etl['tiempo_libre'] = 24 - (df_etl['study_hours_per_day'] + df_etl['social_media_hours'] + df_etl['netflix_hours'] + df_etl['sleep_hours'] + df_etl['part_time_job'] * 4)
df_etl['ratio_estudio_sueno'] = df_etl['study_hours_per_day'] / df_etl['sleep_hours']

# Clasificaci√≥n de rendimiento acad√©mico
df_etl['rendimiento_academico'] = pd.cut(df_etl['exam_score'],
                                         bins=[0, 50, 75, 100],
                                         labels=['Bajo', 'Medio', 'Alto'])

# Guardar como Excel
output_excel_path = "data/student_data_etl_limpio.xlsx"
df_etl.to_excel(output_excel_path, index=False)

output_excel_path
```

```python
# =========================================
# IMPORTACI√ìN DE LIBRER√çAS
# =========================================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

sns.set(style="whitegrid")
palette_soft = ['#08306B', '#08519C', '#2171B5', '#4292C6', '#6BAED6', '#9ECAE1']

# =========================================
# CARGA DE DATOS
# =========================================
df = pd.read_excel("data/student_data_etl_limpio.xlsx")

# =========================================
# 1Ô∏è‚É£ CLUSTERING DE PERFILES ESTUDIANTILES
# =========================================

# Selecci√≥n de variables relevantes
features_cluster = ['study_hours_per_day', 'sleep_hours', 'exercise_frequency',
                    'mental_health_rating', 'ratio_estudio_sueno', 'social_media_hours', 'netflix_hours']

X_cluster = df[features_cluster].copy()

# Estandarizaci√≥n
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

# Determinar n√∫mero √≥ptimo de cl√∫sters (elbow)
wcss = []
for i in range(2, 10):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(2, 10), wcss, marker='o', color=palette_soft[1])
plt.title("M√©todo del Codo - Selecci√≥n de K")
plt.xlabel("N√∫mero de Cl√∫sters")
plt.ylabel("WCSS")
plt.show()

# Ajustar modelo final con K=3 (puedes cambiarlo)
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(X_scaled)

# Visualizaci√≥n con PCA
pca = PCA(n_components=2)
pca_components = pca.fit_transform(X_scaled)
df['PCA1'], df['PCA2'] = pca_components[:, 0], pca_components[:, 1]

sns.scatterplot(data=df, x='PCA1', y='PCA2', hue='cluster', palette=palette_soft)
plt.title("Visualizaci√≥n de Cl√∫sters de Estudiantes")
plt.show()

# Silhouette Score
score = silhouette_score(X_scaled, df['cluster'])
print(f"üß† Silhouette Score del Clustering: {score:.2f}")

# =========================================
# 3Ô∏è‚É£ CLASIFICADOR DE RENDIMIENTO ACAD√âMICO
# =========================================

# Encoding de target
le = LabelEncoder()
df['rendimiento_encoded'] = le.fit_transform(df['rendimiento_academico'])

# Features para clasificaci√≥n
features_class = ['study_hours_per_day', 'sleep_hours', 'exercise_frequency',
                  'mental_health_rating', 'ratio_estudio_sueno', 'social_media_hours', 'netflix_hours', 'part_time_job',
                  'extracurricular_participation', 'diet_quality', 'internet_quality']

X_class = df[features_class]
y_class = df['rendimiento_encoded']

# Divisi√≥n entrenamiento / prueba
X_train, X_test, y_train, y_test = train_test_split(X_class, y_class, test_size=0.2, random_state=42)

# Modelo Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Evaluaci√≥n
print("\nüéØ Reporte de Clasificaci√≥n:")
print(classification_report(y_test, y_pred, target_names=le.classes_))

# Matriz de Confusi√≥n
conf_mat = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_mat, annot=True, fmt='d', cmap="PuBu", xticklabels=le.classes_, yticklabels=le.classes_)
plt.title("Matriz de Confusi√≥n - Clasificaci√≥n de Rendimiento")
plt.xlabel("Predicci√≥n")
plt.ylabel("Real")
plt.show()

# Importancia de Variables
importances = pd.Series(model.feature_importances_, index=features_class).sort_values(ascending=True)
importances.plot(kind='barh', color=palette_soft[3])
plt.title("Importancia de Variables - Random Forest")
plt.show()

# =========================================
# 4Ô∏è‚É£ AN√ÅLISIS DE TIEMPO LIBRE VS BURNOUT
# =========================================

# Categor√≠as de tiempo libre
df['nivel_tiempo_libre'] = pd.cut(df['tiempo_libre'],
                                  bins=[-1, 4, 8, 24],
                                  labels=['Muy Bajo', 'Medio', 'Alto'])

# Boxplot: salud mental seg√∫n tiempo libre
sns.boxplot(data=df, x='nivel_tiempo_libre', y='mental_health_rating', palette=palette_soft)
plt.title("Salud Mental seg√∫n Nivel de Tiempo Libre")
plt.xlabel("Nivel de Tiempo Libre")
plt.ylabel("Puntaje de Salud Mental")
plt.show()

# Boxplot: rendimiento acad√©mico seg√∫n tiempo libre
sns.boxplot(data=df, x='nivel_tiempo_libre', y='exam_score', palette=palette_soft)
plt.title("Puntaje de Examen seg√∫n Tiempo Libre")
plt.xlabel("Nivel de Tiempo Libre")
plt.ylabel("Puntaje en Examen")
plt.show()
```

üî∑ 1Ô∏è‚É£ CLUSTERING DE PERFILES ESTUDIANTILES
‚úÖ Objetivo:
Agrupar estudiantes con comportamientos similares en base a h√°bitos y estilo de vida.

üîç Proceso paso a paso:
Selecci√≥n de variables clave:

study_hours_per_day, sleep_hours, exercise_frequency, mental_health_rating, ratio_estudio_sueno, social_media_hours, netflix_hours

C√°lculo del n√∫mero √≥ptimo de cl√∫sters (k):

Se aplic√≥ el m√©todo del codo (elbow method) para ver cu√°ntos grupos son ideales.

Tambi√©n se calcul√≥ el Silhouette Score, que mide qu√© tan bien definidos est√°n los cl√∫sters.
